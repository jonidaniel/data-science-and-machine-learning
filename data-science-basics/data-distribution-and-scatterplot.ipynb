# Data Distribution and Scatterplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Uniform Data Distribution

```python
# Data distribution (uniform)
# Create an array containing random floats
data = np.random.uniform(0.0, 5.0, 10000) # Low, high, size

# Draw a histogram from the distribution
plt.hist(data, 100) # Data, num of bars
plt.show()
```

### Normal Data Distribution

```python
# Normal data distribution (Gaussian)
# Create an array containing floats concentrated around a value
data = np.random.normal(5.0, 1.0, 10000) # Mean, std, size

# Draw a histogram from the distribution
plt.hist(data, 100) # Data, num of bars
plt.show()
```

### Scatterplot

```python
# Draw a scatter plot

# X and y need to be of the same length
X = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

plt.scatter(X, y)
plt.grid() # Configure a grid
plt.show()
```
