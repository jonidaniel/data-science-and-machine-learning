### Linear Regression Model
#### y = mx + b, where m = slope & b = y-intercept

###### slope, coefficient = kaltevuuskulma, kerroin
###### intercept = nollakohta
###### scatter plot = hajontakuvio, sirontakuvio

```python
from sklearn import datasets # For sample dataset
import seaborn as sns # For scatterplotting
from sklearn.model_selection import train_test_split # For data splitting
from sklearn import linear_model # For the linear regression model
from sklearn.metrics import mean_squared_error, r2_score # For the model's performance
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes.feature_names
```

```python
X = diabetes.data
Y = diabetes.target
```

```python
X.shape, Y.shape
```

#### 80/20 Data Split

##### 80 % of the data is used as a training set to build the model
##### Then the built model is assigned to make the prediction on the 20 %

```python
# Do the split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

#### Build the model

```python
model = linear_model.LinearRegression()
```

```python
# Build the training model
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

#### Prediction results

```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
# Use the modulo operator to round numbers off to 2 decimals
print("Mean squared error (MSE): %.2f" % mean_squared_error(Y_test, Y_pred))
print("Coefficient of determination (R^2): %.2f" % r2_score(Y_test, Y_pred))
```

#### Display a scatterplot

```python
sns.scatterplot([Y_test, Y_pred], alpha = 0.5) # Use alpha parameter to make the markers more translucent
```

```python
sns.scatterplot(Y_test, marker = "+") # Use marker parameter to modify the markers
```
