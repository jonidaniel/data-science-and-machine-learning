# Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
# Draw a line of polynomial regression

X = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100]

# Make a polynomial model
myModel = np.poly1d(np.polyfit(X, y, 3)) # x values, y values, deg (degree of the fitting polynomial)

# Specify how the line will display
myLine = np.linspace(1, 22, 100) # Start from position 1, end at position 22, 100 positions along the way

# Draw the original scatter plot
plt.scatter(X, y)
# Draw the line of polynomial regression
plt.plot(myLine, myModel(myLine))
plt.show()
```

```python
# R-squared
# Ranges from 0 to 1, where 0 means no relationship, and 1 means 100% related
# How is the relationship between the x-axis and y-axis values?
# I.e., how well does my data fit in a polynomial regression?

X = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100]

# Make a polynomial model
myModel = np.poly1d(np.polyfit(X, y, 3)) # x values, y values, deg (degree of the fitting polynomial)

r2_score(y, myModel(X))
```
